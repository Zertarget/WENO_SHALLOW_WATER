```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import csv

with open('heights_2.csv', 'r', encoding='UTF-8') as file:
    data_str = list(csv.reader(file))
X = [float(i) for i in data_str[0][1:-1]]
H = [[float(i) for i in string[1:-1]] for string in data_str[1:-1]]
with open('speeds_2.csv', 'r', encoding='UTF-8') as file:
    data_str = list(csv.reader(file))
U = [[float(i) for i in string[1:-1]] for string in data_str[1:-1]]

fig = plt.figure()
p = fig.add_subplot(111)

p.set_xlim(0, X[-1] - X[0] + (X[1] - X[0]))
p.set_ylim(max(0.0, min([min(data_part) for data_part in U])) - 1,
           max(max([max(data_part) for data_part in U]), max(H[0])) + 1)

# set up empty lines to be updates later on
l1, = p.plot([], [], 'b')
l2, = p.plot([], [], 'r')


def run(i):
    plt.title("time = " + data_str[i + 1][0][0:5])

    l1.set_data(X, H[i])
    l2.set_data(X, U[i])
    return []


animation.FuncAnimation(fig,
                        run,
                        frames=len(H),
                        interval=20).save('sample_data/res_1.gif', writer="pillow")
```
